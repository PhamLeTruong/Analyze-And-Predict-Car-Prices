# Đọc dữ liệu

```python
df = pd.read_csv('/content/drive/MyDrive/CV/Analyze-And-Predict-Car-Price/Source_Code/final_data.csv')
df
```

# Mã hóa dữ liệu (Encoding)

```python
df_encode = pd.DataFrame()
```

```python
def label_encoder(category_col):
    encoder = LabelEncoder()
    enc = encoder.fit_transform(df[category_col])
    return list(enc)
```

```python
for col_name in df.columns:
    encode = df[col_name]
    if type(df[col_name][0])==str:
        if len(df[col_name].unique()) > 2:
            encode = pd.get_dummies(df[col_name], prefix=col_name, dtype=int)
        else:
            encode = pd.DataFrame(label_encoder(col_name), columns=[col_name])
    df_encode = pd.concat([df_encode, encode], axis=1)

```

```python
df_encode
```

# Vector hóa dữ liệu

```python
X = df_encode[df_encode.columns[:-1]].values
y = df_encode['Price'].values
```

```python
X
```

```python
y
```

# Chia tập dữ liệu Train-Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
print(f"Kích thước tập train là: {X_train.shape}")
print(f"Kích thước tập test là: {X_test.shape}")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_temp = scaler.fit_transform(X_train)
X_test_temp = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=23, random_state=0) # số lượng đặc trưng ban đầu của dữ liệu là 23
X_train_temp = pca.fit_transform(X_train_temp)
X_test_temp = pca.transform(X_test_temp)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import time

eval_metrics = {}

def predict_model(model, X_train_temp, y_train, X_test_temp):
  start_time = time.time()
  model.fit(X_train_temp, y_train)
  y_pred = model.predict(X_test_temp)
  return y_pred, round(time.time()-start_time,2)

def eval_model(y_test, y_pred):
  return int(mean_squared_error(y_test, y_pred, squared=False)) , int(mean_absolute_error(y_test, y_pred)), round(r2_score(y_test, y_pred),2)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

y_pred, total_time = predict_model(LinearRegression(), X_train_temp, y_train, X_test_temp)
rmse, mae, r2 = eval_model(y_test, y_pred)
eval_metrics['Linear Regression'] = [rmse, mae, r2, total_time]
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

y_pred, total_time = predict_model(DecisionTreeRegressor(random_state=0), X_train_temp, y_train, X_test_temp)
rmse, mae, r2 = eval_model(y_test, y_pred)
eval_metrics['Decision Tree'] = [rmse, mae, r2, total_time]
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor

y_pred, total_time = predict_model(KNeighborsRegressor(), X_train_temp, y_train, X_test_temp)
rmse, mae, r2 = eval_model(y_test, y_pred)
eval_metrics['K-nearest neighbors'] = [rmse, mae, r2, total_time]
```

# Support Vector Machine

```python
from sklearn.svm import SVR

y_pred, total_time = predict_model(SVR(), X_train_temp, y_train, X_test_temp)
rmse, mae, r2 = eval_model(y_test, y_pred)
eval_metrics['Support Vector Machine'] = [rmse, mae, r2, total_time]
```

# So sánh giữa các mô hình

```python
pd.DataFrame(eval_metrics, index=['rmse', 'mae', 'r2_score', 'total_time'])
```

# Xây dựng Pipeline mô hình máy học
Xây dựng Pipeline kết hợp với GridSearchCV để tối ưu mô hình Linear Regression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
import time

start_time = time.time()
steps = [
    ('scaler', MinMaxScaler()), # chuẩn hóa dữ liệu
    ('pca', PCA(random_state=0)),   # giảm chiều dữ liệu
    ('linear_regression', LinearRegression())   # mô hình dữ liệu
]

# Tạo pipeline
pipeline = Pipeline(steps)

# Thiết lập các tham số cần tìm kiếm
param_grid = {
    'pca__n_components': [i+1 for i in range(X_train.shape[1])],  # Số lượng feature từ 1 đến 412
    'pca__svd_solver': ['auto', 'full'],
    'linear_regression__fit_intercept': [True, False],
    'linear_regression__copy_X': [True, False]
}

# Tạo đối tượng GridSearchCV với pipeline
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='r2', verbose=1)

# Tiến hành huấn luyện trên tập train
grid_search.fit(X_train, y_train)

# In ra mô hình tốt nhất được huấn luyện
model = grid_search.best_estimator_
print("Best model: ", model)

# Tổng thời gian
total_time = time.time() - start_time
print(f'Tổng thời gian: {int(total_time)} giây')
```

# Đánh giá mô hình máy học

```python
from sklearn.metrics import r2_score

y_pred, total_time = predict_model(model, X_train, y_train, X_test)
rmse, mae, r2 = eval_model(y_test, y_pred)
eval_metrics['Linear_Regression_Optimized'] = [rmse, mae, r2, total_time]
```

```python
pd.DataFrame(eval_metrics, index=['rmse', 'mae', 'r2_score', 'total_time'])
```
